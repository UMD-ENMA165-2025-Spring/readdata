# Defining a Plotting Function

## Import Packages

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Define Function

```python
def plotregion(region_path):
    """graph of gdp per capita vs year

    - region_path: path to data in csv file
    - no return value
    """
    variable0 = np.loadtxt(skiprows=1,
                              fname=region_path,
                              usecols=range(1,13),
                              delimiter=",")
    f = open(region_path,"r")
    s = f.readline()
    f.close()
    row = s.strip("\n")
    
    values = row.split(",")
    
    years = []
    for gdp in values[1:13]:
        gdp = gdp.strip("gdpPercap_")
        years.append(int(gdp))

    f = open(region_path,"r")
    s = f.read()
    f.close()
    countries = []
    for row in s.strip().split("\n"):
        values = row.split(",")
        countries.append(values[0])
    countries = countries[1:]

    for line in zip(countries,variable0):
        plt.plot(years,line[1], label=line[0])
    plt.legend()
    plt.ylabel("gdp/capita")  
    plt.xlabel("year")
```

```python
plotregion("data/gapminder_gdp_africa.csv")
```

```python
plotregion("data/gapminder_gdp_americas.csv")
```
